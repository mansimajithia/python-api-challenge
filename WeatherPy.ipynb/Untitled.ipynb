```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
from sklearn import datasets


# Import API key
from config import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"
# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
output_data_file
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
count = 0
city_names = []
set_count = 1
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
executing = ""
units = 'imperial'
base_url = "http://api.openweathermap.org/data/2.5/weather?"

print("Beginning Data Retrieval")
print("---------------------------")
for city in cities:
    try: 
#         executing = "Building querty"
        query_url = f"{base_url}appid={weather_api_key}&q={city}&units={units}"
#     base_url + "appid=" + weather_api_key + "&q=" + city + "units"

#         executing = "request.get"
        response = requests.get(query_url)
        
#         executing = "response.json"
        weather = response.json()
#         pprint(weather)
        
#         executing = "appending city"
        city_names.append(weather['name'])

        
#         executing = "appending cloudiness"
        cloudiness.append(weather['clouds']['all'])
        
#         executing = "appending country"
        country.append(weather['sys']['country'])
        
#         executing = "appending date"
        date.append(weather['dt'])
        
#         executing = "appending humidity"
        humidity.append(weather['main']['humidity'])
        
#         executing = "appending lat"
        lat.append(weather['coord']['lat'])
        
#         executing = "appending lng"
        lng.append(weather['coord']['lon']) 
        
#         executing = "appending max temo"
        max_temp.append(weather['main']['temp_max'])
        
#         executing = "appending wind speed"
        wind_speed.append(weather['wind']['speed'])

        
        count += 1

        print(f"Processing record {count} of {set_count} | {city}")

        if count >= 50:
            set_count += 1
            count = 0
    except:
#         print (f"error: {executing}")
        print("City not found. Skipping...") 
```

```python
weather_data = {
     "City": city_names,
     "Cloudiness": cloudiness,
     "Country": country,
     "Date": date,
     "Humidity": humidity,
     "Lat": lat,
     "Lng": lng,
     "Max Temp": max_temp,
     "Wind Speed": wind_speed    
 }

# Create the data frame and count variables for each columns
weather_df = pd.DataFrame.from_dict(weather_data, orient='index')
weather_df = weather_df.transpose()
weather_df.count()
```

Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_hem['Lat'].astype(float)
y_values = north_hem['Max Temp'].astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red",)
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.title("Max Temp vs Latitude (Northern Hemisphere)")
print(f"The r-squared is: {rvalue}")
plt.show()
```
